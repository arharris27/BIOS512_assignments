```r
turbine = read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv')
turbine |> head()
    
```

Each row in the table represents a single turbine.

Make a bar chart üìä **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
#Making a variable of turbine count to get a better idea of turbine frequency by province.

turbine |>
    group_by(province_territory) |>
    mutate(turbine_count=n()) |>
    head()
```

```r
options(repr.plot.height = 4, repr.plot.width = 6, repr.plot.res = 250)

turbine <- turbine %>%
  mutate(province_territory = fct_infreq(province_territory))

p=ggplot(turbine, aes(y=province_territory)) + 
    geom_bar() +
    labs(title="Turbine Count by Province/Territory", x="Turbine Frequency", y="Province/Territory")
p
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
options(repr.plot.height = 4, repr.plot.width = 6, repr.plot.res = 250)

turbine_lumped = turbine |>
    mutate(manu_lumped = manufacturer |> 
               fct_lump(n = 5) |> 
               fct_infreq() |> 
               fct_relevel('Other', after= Inf) 
               |> fct_rev())

p = ggplot(turbine_lumped, aes(y = manu_lumped)) +
    geom_bar() +
    labs(title='Turbine Count by Manufacturer', x = 'Turbine Frequency', y = 'Manufacturer')
           
p
```
